```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptops.csv')
df
```

```python
df.info()
```

```python
#data cleaning

def clean_price_column(price):  #isinstance function returns true if the specified object is of the specified type,
    if isinstance(price,str): # replace() method replaces a soecified phrase with another specified 
        price = price.replace(',','')
        if price.isnumeric():
            price = int(price)           # isnumeric() method returns True if all the character are numeric (0-9), other
            return price
        if len(price) == 0:
            return np.nan  #np.nan = not a number
```

```python
clean_price_column('39,990')
```

```python
df['Cleaned_Price'] = df['price'].apply(clean_price_column)
df
```

```python
df.drop(columns='price',inplace=True)
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        rating = rating.split()[0]
        return float(rating)
    return rating
    
```

```python
df['Cleaned_Rating'] = df['rating'].apply(clean_rating_column)
df
```

```python
df.drop(columns='rating',inplace=True)
df
```
